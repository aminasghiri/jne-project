```python
import pandas as pd

# Exemple d'extrait du dataset tabulaire AI-ready (5 lignes, 6 colonnes)
data_example = {
    "timestamp": [
        "2025-01-01 08:00", "2025-01-01 08:15", "2025-01-01 08:30",
        "2025-01-01 08:45", "2025-01-01 09:00"
    ],
    "room_temp_C": [21.5, 21.7, 21.8, 21.6, 21.4],
    "humidity_int_%": [45, 46, 47, 46, 45],
    "CO2_ppm": [620, 645, 670, 660, 640],
    "occupancy": [12, 14, 15, 15, 14],
    "power_HVAC_kW": [3.2, 3.4, 3.5, 3.3, 3.1]
}
df_example = pd.DataFrame(data_example)

# Tableau des caractéristiques des variables (statique + dynamique)
variables_info = {
    "Variable": [
        "Surface (m²)", "Volume (m³)", "Orientation (°)", "Capacité d’occupation",
        "Température intérieure (°C)", "Humidité relative intérieure (%)", "CO₂ (ppm)",
        "Luminosité (lux)", "Occupation (nb pers.)", "Setpoint HVAC (°C)", 
        "Etat HVAC (ON/OFF)", "Puissance HVAC (kW)", 
        "Température extérieure (°C)", "Humidité extérieure (%)",
        "Irradiance solaire (W/m²)", "Vitesse du vent (m/s)"
    ],
    "Source": [
        "BIM", "BIM", "BIM", "BIM",
        "IoT capteur", "IoT capteur", "IoT capteur",
        "IoT capteur", "IoT capteur", "BMS",
        "BMS", "BMS",
        "Météo API", "Météo API", "Météo API", "Météo API"
    ],
    "Type": [
        "Statique", "Statique", "Statique", "Statique",
        "Dynamique", "Dynamique", "Dynamique",
        "Dynamique", "Dynamique", "Dynamique",
        "Dynamique", "Dynamique",
        "Dynamique", "Dynamique", "Dynamique", "Dynamique"
    ],
    "Fréquence": [
        "-", "-", "-", "-",
        "15 min", "15 min", "15 min",
        "15 min", "15 min", "15 min",
        "15 min", "15 min",
        "15 min", "15 min", "15 min", "15 min"
    ]
}
df_variables = pd.DataFrame(variables_info)

# Export vers Word
from docx import Document

doc = Document()
doc.add_heading("Section 4.2 – AI-ready Dataset", level=1)

# Ajouter figure (extrait dataset tabulaire)
doc.add_heading("Figure 10: Exemple d’extrait du dataset tabulaire AI-ready", level=2)
t1 = doc.add_table(df_example.shape[0]+1, df_example.shape[1])
t1.style = "Table Grid"

# Header
for j, col in enumerate(df_example.columns):
    t1.cell(0, j).text = col
# Contenu
for i in range(df_example.shape[0]):
    for j in range(df_example.shape[1]):
        t1.cell(i+1, j).text = str(df_example.iat[i, j])

doc.add_paragraph("")

# Ajouter tableau des variables
doc.add_heading("Tableau 5: Caractéristiques des variables du dataset AI-ready", level=2)
t2 = doc.add_table(df_variables.shape[0]+1, df_variables.shape[1])
t2.style = "Table Grid"

# Header
for j, col in enumerate(df_variables.columns):
    t2.cell(0, j).text = col
# Contenu
for i in range(df_variables.shape[0]):
    for j in range(df_variables.shape[1]):
        t2.cell(i+1, j).text = str(df_variables.iat[i, j])

# Sauvegarde
output_path = "/mnt/data/Section_4.2_AI_ready_dataset.docx"
doc.save(output_path)
output_path
```
